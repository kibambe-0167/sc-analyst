```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./courier_costs.csv")
df
```

```python
plt.figure(figsize=(6,3))
sns.relplot(data=df, x="lmr", y="cost",
            col="delivery_type", 
            hue="courier_id",
            kind='scatter'
            )
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.catplot(data=df, x="delivery_type", y="cost", hue="courier_id", kind="bar")
plt.show()
```
