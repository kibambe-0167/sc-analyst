```python
plt.figure(figsize=(6, 4))
sns.countplot(
    data=df,
    x="HubID",
    hue="CourierId",
    # order=df["LMR"].value_counts().index,  # for sorting
)
plt.title("Counts of HubID")
plt.xticks(rotation=90)
plt.show()
```
